```python
import pandas as pd
import joblib

# Load model
bundle = joblib.load("isolation_forest_fixed.pkl")

model = bundle["model"]
scaler = bundle["scaler"]
features = bundle["features"]
threshold = bundle["threshold"]

manual_data = pd.DataFrame([{
    "Vehicle_Speed_kmh": 0,
    "Vehicle_Speed_RPM_mean": 0,
    "Wheel_RPM_std": 0,
    "Battery_Voltage_V": 0,
    "Battery_Current_A": 0,
    "Battery_Power_W": 0,
    "Power_per_Speed":0,
    "Battery_SOC_Pct": 0,
    "Battery_Resistance_Ohms": 0,
    "Battery_Temp_Max_C": 0,
    "Temp_Stress_Index": 0,
    "Throttle_Pct": 0,
    "Torque_Command_Error": 0,
    "Brake_Pedal_Pct": 0,
    "Brake_Pressure_Front_kPa": 0,
    "Brake_Pressure_Rear_kPa": 0,
    "Brake_Pressure_Diff": 0
}])

# Ensure correct order
manual_data = manual_data[features]

# Scale
X_scaled = scaler.transform(manual_data)

# Predict
score = -model.score_samples(X_scaled)[0]
anomaly = int(score > threshold)

print("Score:", score)
print("Anomaly:", anomaly)
```
